# Revisit the Preprocessing: Scale the data

After revisiting past exercises, I believe that scaling the data will help improve the 
Logistic Regression model scores but will not affect the Random Forest Classifier Model much to be honest.
We'll see if I'm right.


```python
#Scale the data
#======================================================================================
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(train_XD, train_Y, random_state=25)
scaler = StandardScaler().fit(X_train)
train_scaled_X = scaler.transform(X_train)
test_scaled_X = scaler.transform(X_test)
```

```python
#Train the Logistic Regression model on the scaled data and print the model score
#=================================================================================
classifier = LogisticRegression(max_iter=10000)
classifier
classifier.fit(train_scaled_X, Y_train)
print(f"Training Data Score: {classifier.score(train_scaled_X, Y_train)}")
print(f"Testing Data Score: {classifier.score(test_scaled_X, Y_test)}")
```

```python
#Train a Random Forest Classifier model on the scaled data and print the model score
#====================================================================================
clf = RandomForestClassifier(random_state=28, n_estimators=500).fit(train_scaled_X, Y_train)
print(f"Training Score: {clf.score(train_scaled_X, Y_train)}")
print(f'Testing Score: {clf.score(test_scaled_X, Y_test)}')
```

# In conclusion:
